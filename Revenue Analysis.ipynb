```python
import pandas as pd

from urllib.parse import urlparse
from urllib.parse import parse_qs

from datetime import date
```

```python
df = pd.read_csv("C:/Users/reshma.v.kotwani/Downloads/data.tsv",sep='\t')
```

```python
df
```

```python
df = df.groupby('ip').filter(lambda g: (g['event_list'] == 1.0).any())
```

```python
df = df.drop(['hit_time_gmt', 'geo_city', 'geo_region', 'geo_country','pagename', 'page_url'], axis=1)
```

```python
g = df.groupby('ip')

df_out = (pd.concat([g.head(1), g.tail(1)]).sort_values(['ip','date_time'], ascending=False).reset_index(drop=True))
```

```python
df_out
```

```python
mask = df_out['event_list'] == 1.0
```

```python
product = df_out[mask]
search = df_out[~mask]
```

```python
search
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
search['Search Engine Domain'] = search['referrer'].apply(urlparse).apply(lambda x: x.netloc[4:])
```

```python
search['Search Keyword'] = search['referrer'].apply(urlparse).apply(lambda x: parse_qs(x.query)['q'][0])
```

```python
search
```

```python
product = pd.DataFrame(product['product_list'].str.split(',').tolist(), index=product.ip).stack()
```

```python
product = product.reset_index([0, 'ip'])
```

```python
product
```

```python
new = product[0].str.split(";", expand = True)
```

```python
product["Category"]= new[0]
product["Product Name"]= new[1]
product["Num of Items"]= new[2]
product["Revenue"]= new[3]
product["Custom Event"]= new[4]
#product["eVar"]= new[5]
product.drop(columns =[0], inplace = True)
```

```python
cols = ['Num of Items','Revenue']
product[cols] = product[cols].apply(pd.to_numeric)
```

```python
total_revenue = product.groupby('ip')['Revenue'].sum()
total_revenue
```

```python
df_final = search.merge(total_revenue, on='ip', how='inner').sort_values(['Revenue'], ascending=False)
```

```python
df_final
```

```python
df_final = df_final.drop(['date_time','user_agent','ip','event_list','product_list','referrer'],axis=1)
df_final
```

```python
today = date.today()
filename = str(today) + '_SearchKeywordPerformance.tab'
print(filename)
```

```python
df_final.to_csv(filename, sep="\t")
```
